```python
def bubble_sort(nums):  
    # Устанавливаем swapped в True, чтобы цикл запустился хотя бы один раз
    swapped = True
    while swapped:
        swapped = False
        for i in range(len(nums) - 1):
            if nums[i] > nums[i + 1]:
                # Меняем элементы
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
                # Устанавливаем swapped в True для следующей итерации
                swapped = True


```

```python
import time
import memory_profiler
from numpy import random




yAxisMemory = []
yAxisTime = []
xAxis = []

i=0
arr=random.randint(10000, size=(1000))
while i < 500:
    x=arr
    xAxis.append(i)
    m1 = memory_profiler.memory_usage()
    
    start_time = time.time()
    bubble_sort(x)
    yAxisTime.append(time.time() - start_time)
    m2=memory_profiler.memory_usage() 
    yAxisMemory.append(m2[0] - m1[0])
    #print(f"It took {yAxisTime[i]} Secs and {yAxisMemory[i]} Mb ")
    print(i)
    i=i+1
    


```
